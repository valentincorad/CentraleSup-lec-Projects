## Résumé

Dans un premier temps les variables d'intérêt ont été étudiées dans les deux datasets.  Les deux datasets identifient par ligne un batiment unique avec leurs caractéristiques techniques. La variable permettant d'identifier un bâtiment est 'OSEBuildingID'. Cependant les deux datasets n'identifient pas le même nombre de bâtiments unique entre 2015 et 2016. 

Certaines variables ont été retirées afin d'avoir les mêmes variables dans les deux datasets et ainsi pouvoir les concaténer.

Par la suite, le dataset obtenu a été aggrégé par la moyenne suivant l'identifiant 'OSEBuildingID' pour n'obtenir que des bâtiments uniques par ligne. Le dataset formé contient 3432 batiments uniques et 44 variables associées issus des relevés 2015 et 2016 ainsi que les 148 autres batiments non communs aux deux datasets simultanément. 

Dans un deuxième temps un travail sur la variables a été effectué, notamment un retrait des nan dans les targets, un retrait des doublons, retrait des valeurs aberrantes,...

Les variables renseignées à moins de 50% ont été retirées ainsi que les variables très corrélées (variable énergétiques similaire avec différente unité). Il en est de même avec les doublons et les lignes nulles.

Enfin les distributions des targets ont été visualisés pour retirer les outliers selon la méthode impliquant l'utilisation de l'écart interquartile (IQR) et ce après une transformation (Power Transformation) pour obtenir une distribution normale.

Enfin l'analyse des variables catégorielles a permis de mettre en évidence que la plupart des bâtiments correspondent à des bâtiments familiaux et non résidentiels et sont situés dans le centre et dans le district 7.

Toutes les variables catégorielles contiennent des valeurs faisant référence à des batiments d'habitation, en particulier ceux contenant le mot 'family'. Ex :"Multifamily Housing" Toutes les lignes contenant ce mot ont été retirées.

Après ce nettoyage il reste 763 bâtiments et 26 variables. Les principaux bâtiments consommant le plus d'énergie sont ceux relatifs à l'éducation (Campus + SPS District K12), à des grands bureaux (large office) et des hopitaux (medical office). Ils sont situés dans le centre et la zone de Lake Union.

```python
# Fonction permettant d'identifier les variables non communes à deux dataframs

def diff(list1, list2):
    c = set(list1).union(set(list2))  # or c = set(list1) | set(list2)
    d = set(list1).intersection(set(list2))  # or d = set(list1) & set(list2)
    return list(c - d)
```

```python
# Création d'une fonction permettant de visualiser le taux de remplissage par variable

def prop(df, figsize):
    
    # Affichage de la forme du dataset
    print("Le dataset contient {} lignes et {} colonnes.".format(df.shape[0], df.shape[1]))
    
    # Proportion de valeurs nulles
    prop_nan_df = df.isna().sum().sum()/(df.shape[0]*df.shape[1])*100
    print('Il y a {:.2f}% de NaN.'.format(prop_nan_df))
    
    # Visualisation des valeurs non nulles
    df_null=(df.notnull().sum()/df.shape[0]).sort_values(ascending=False)*100
    
    # Création de la figure
    plt.figure(figsize=figsize)
    plt.style.use('ggplot')
    ax=sns.barplot(x=df_null.values, y=df_null.index, orient='h')
    plt.title('Proportion de valeurs présentes par colonne')
    plt.xticks(rotation=0)
    plt.yticks(fontsize=14)
    plt.ylabel("Nom de la colonne")
    plt.xlabel('Pourcentage de valeurs présentes (%)')

    # Affichage de la valeur à côté des barres
    for index, value in enumerate(df_null.values):
        plt.text(value, index+0.4, str(value.round(2)))

    plt.show()
```

Logarithmic transformation is a convenient means of transforming a highly skewed variable into a more normalized dataset. When modeling variables with non-linear relationships, the chances of producing errors may also be skewed negatively. In theory, we want to produce the smallest error possible when making a prediction, while also taking into account that we should not be overfitting the model. 

Overfitting occurs when there are too many dependent variables in play that it does not have enough generalization of the dataset to make a valid prediction. Using the logarithm of one or more variables improves the fit of the model by transforming the distribution of the features to a more normally-shaped bell curve.

https://dev.to/rokaandy/logarithmic-transformation-in-linear-regression-models-why-when-3a7c

#### Log transformation des targets

```python
# Map data to a normal distribution (Log transformation)
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method='yeo-johnson', standardize=True)
pt.fit(df_concat_complete[targets])
df_targets_log = pd.DataFrame(pt.transform(df_concat_complete[targets]), columns = 'log' + df_concat_complete[targets].columns, index = df_concat_complete.index)


# Distribution et boxplot des targets après log transformation et standardisation
plt.style.use('bmh')

fig = plt.figure(1, figsize=(14, 8))
for i,label in enumerate(["logTotalGHGEmissions", "logSiteEnergyUse(kBtu)"]) :
    plt.subplot(2,2,i+1)
    sns.distplot(df_targets_log[label], bins=int(1 + np.log2(len(df_targets_log))), color='steelblue')
    plt.subplot(2,2,i+3)
    sns.boxplot(df_targets_log[label], orient="h")
plt.show()
```

```python
# Dataframe avec targets transformées
df_log = pd.concat([df_concat_complete, df_targets_log], axis=1)
```

#### Retrait des outliers des targets

Le retrait des outliers est effectué avec l'utilisation de l'écart interquartile (IQR). Les valeurs outliers sont situées hors des "moustaches" c'est à dire au-delà de Q3 + 1.5*(Q3-Q1) et au-deçà de Q1 - 1.5*(Q3-Q1).
Or pour pouvoir appliquer ce principe, la distribution de la variable doit se rapprocher d'une distribution normale, c'est pourquoi une transformation ( log transformation via PowerTransformer) est appliqué au préalable.

```python
# Calcul des quartiles et IQR pour l'énergie
Q1 = np.percentile(df_log["logSiteEnergyUse(kBtu)"], 25)
Q3 = np.percentile(df_log["logSiteEnergyUse(kBtu)"], 75)

born_sup_energy = Q3 + 1.5*(Q3-Q1)
born_inf_energy = Q1 - 1.5*(Q3-Q1)
```

```python
# Calcul des quartiles et IQR pour le CO2
Q1 = np.percentile(df_log["logTotalGHGEmissions"], 25)
Q3 = np.percentile(df_log["logTotalGHGEmissions"], 75)

born_sup_co2 = Q3 + 1.5*(Q3-Q1) 
born_inf_co2 = Q1 - 1.5*(Q3-Q1)
```

```python
# Retrait des outliers hors des IQR
df_log = df_log[(df_log["logSiteEnergyUse(kBtu)"] < born_sup_energy)
            & (df_log["logSiteEnergyUse(kBtu)"] > born_inf_energy)]

df_log = df_log[(df_log["logTotalGHGEmissions"] < born_sup_co2)
            & (df_log["logTotalGHGEmissions"] > born_inf_co2)]
```

```python
# Visualisation de la distribution des targets après retrait des outliers
plt.style.use('bmh')

fig = plt.figure(1, figsize=(14, 8))
for i,label in enumerate(["logTotalGHGEmissions", "logSiteEnergyUse(kBtu)"]) :
    plt.subplot(2,2,i+1)
    sns.distplot(df_log[label], bins=int(1 + np.log2(len(df_log))), color='steelblue')
    plt.subplot(2,2,i+3)
    sns.boxplot(df_log[label], orient="h")
plt.show()
```

Dans cette partie les indicateurs sont analysées. Il s'est avéré qu'il était plus judicieux d'analyser certains uniquement par continents, d'autres uniquement par pays et parfois les deux.

Trois types de graphiques (boxplot, histogramme et lineplot) et une carte choroplèthe ont été crées. 

Les histogrammes crées prennent en compte l'année 2015. Les boxplots prennent en compte l'ensemble de la période de temps étudié allant de 1999 à 2015. ### 3.1 Méthodologie

Création des fonctions permettant de générer les graphiques pour l'analyse par pays ou par continents ainsi que les cartes choroplèthes

```python
# Dataset continent et pays
df_cont=df[df['Country Name'].isin(liste_continents)]
df_pays=df[df['Country Name'].isin(liste_pays)]
```

```python
# Fonction pour un indicateur - analyse par continent

def plot_cont(indicator, title_barplot, xlabel_barplot, title_boxplot,
              xlabel_boxplot, title_lineplot, ylabel_lineplot):

    # Filtrage de dataset continent par indicateur
    df_cont_ind = df_cont[df_cont["Indicator Name"] == indicator]

    # Mise en place d'une condition si l'indicateur analysé n'est pas pris en compte
    if df_cont_ind.empty == True:
        print('Indicateur \'{}\' non pris en compte pour l\'analyse par continent'.format(
            indicator))
    else:
        # Barplot du dataset continent de l'indicateur pour 2015

        plt.figure(figsize=(8, 4))
        plt.style.use('ggplot')
        sns.barplot(x='2015', y='Country Name',
                    data=df_cont_ind.sort_values(by=['2015'], ascending=False))
        plt.title(title_barplot)
        plt.xticks(rotation=0)
        plt.ylabel("Continent")
        plt.xlabel(xlabel_barplot)
        plt.show()

    # Changement d'index du dataset continent - indicateur
        series = df_cont_ind.set_index('Country Name')
        series.head(3)

    # Sélection des colonnes années et transposition
        seriesT = series.loc[:, '1999':'2015'].T

    # Boxplots du dataset continent pour l'indicateur sur l'ensemble de la période de temps étudié : 1999 à 2015
    # Trie des boxplots par leurs médianes

        ordre = seriesT.median()
        ordre.sort_values(ascending=False, inplace=True)
        seriesT_sorted = seriesT[ordre.index]
        plt.figure(figsize=(8, 4))
        plt.title(title_boxplot)
        sns.boxplot(data=seriesT_sorted, orient="h").set(
            xlabel=xlabel_boxplot, ylabel="Continent")

    # Changement d'index du dataset continent - indicateur
        df_cont_ind_set = df_cont_ind.set_index('Country Name')

        plt.figure(figsize=(8, 4))
        for country in df_cont_ind_set.index:
            df_cont_ind_plot = df_cont_ind_set.loc[country, '1999':'2015']
            plt.plot(df_cont_ind_plot.index,
                     df_cont_ind_plot.values, label=country)
        plt.title(title_lineplot)
        plt.xticks(rotation=45)
        plt.ylabel(ylabel_lineplot)
        plt.xlabel("Année")
        plt.style.use('ggplot')
        plt.legend(bbox_to_anchor=(1.05, 1),
                   loc='upper left', borderaxespad=0.)
        plt.show()
```

# 4. Création d'un indicateur économique

Pour synthétiser l'ensemble des indicateurs sélectionnés un indicateur dit "indicateur projet" a été crée et cela uniquement pour l'année 2015.

La création de cet indicateur est basée sur la formule de l'IDH : Indicateur de Développement Humain.

Depuis 2011, la formule suivante est utilisée :

$\text{IDH}=\sqrt[3]{I_{Vie} \times I_{\text{É}ducation} \times I_{Revenu}}$

où$\ I_{Vie}$,$\ I_{\text{É}ducation}$ et$\ I_{Revenu}$, sont respectivement les indices de longévité, le niveau d'éducation et le niveau de revenu.

L'IDH se fonde sur une moyenne géométrique. De plus le logarithme néperien est utilisé sur la composante Revenu. L'objectif est d'éviter qu'un très mauvais score sur l'une des composantes de l'indice puisse être intégralement compensé par un bon résultat sur une autre des composantes.

Enfin chaque variable de la formule correspond à un indice dimensionnel dont la formule est appliqué à chaque indicateur.

${\text{Indice dimensionnel}} = \frac{\text{valeur r}\mathrm{\acute{e}}\text{elle - valeur minimale}}{\text{valeur maximale - valeur minimale}}$



Source : https://fr.wikipedia.org/wiki/Indice_de_d%C3%A9veloppement_humain?veaction=edit&section=4

La formule de l'indicateur projet est la suivante :

$\ {I_{Projet-2015}}=\sqrt[3]{I_{Population} \times I_{GDP} \times I_{Internet}}$

où$\ I_{Population}$,$\ I_{GDP}$ et$\ I_{Internet}$, sont respectivement les indices de la population en âge pour l'éducation secondaire supérieure, le Produit Intérieur Brut et le niveau de couverture internet pour l'année 2015.

Le logarithme néperien est utilisé pour le calcul des indices dimensionnels pour les composantes GDP et Population en âge pour l'éducation secondaire supérieure.

Ce nouvel indicateur sera projeté sur une carte choroplèthe permettant ainsi de visualiser les pays avec le potentiel de client le plus haut.

```python
# Calcul des valeurs minimales et maximales

pop_min=df_pays[df_pays['Indicator Name']=='Population of the official age for upper secondary education, both sexes ''(number)']['2015'].min()
pop_max=df_pays[df_pays['Indicator Name']=='Population of the official age for upper secondary education, both sexes ''(number)']['2015'].max()

internet_min=df_pays[df_pays['Indicator Name']=='Internet users (per 100 people)']['2015'].min()
internet_max=df_pays[df_pays['Indicator Name']=='Internet users (per 100 people)']['2015'].max()

GDP_min=df_pays[df_pays['Indicator Name']=='GDP per capita (current US$)']['2015'].min()
GDP_max=df_pays[df_pays['Indicator Name']=='GDP per capita (current US$)']['2015'].max()
```

```python
# Calcul des indices dimensionnels

I_GDP = (np.log(df_pays[df_pays['Indicator Name'] ==
                        'GDP per capita (current US$)']['2015'])
         - np.log(GDP_min))/(np.log(GDP_max)-np.log(GDP_min))

I_internet = ((df_pays[df_pays['Indicator Name'] ==
                        'Internet users (per 100 people)']['2015'])
         - internet_min)/(internet_max-internet_min)

I_pop = (np.log(df_pays[df_pays['Indicator Name'] ==
                        'Population of the official age for upper secondary education, both sexes ''(number)']['2015'])
         - np.log(pop_min))/(np.log(pop_max)-np.log(pop_min))
```

```python
# Création du dataframe regroupant les trois indices selon le code pays

    # Concatenation avec la colonne 'Contry Code'
df_I_GDP = pd.concat([df_pays['Country Code'], I_GDP], axis=1, join='inner').set_index('Country Code')
df_I_internet = pd.concat([df_pays['Country Code'], I_internet], axis=1, join='inner').set_index('Country Code')
df_I_pop = pd.concat([df_pays['Country Code'], I_pop], axis=1, join='inner').set_index('Country Code')

df_I_GDP.rename(columns={"2015": "I_GDP_2015"}, inplace = True)
df_I_internet.rename(columns={"2015": "I_internet_2015"}, inplace = True)
df_I_pop.rename(columns={"2015": "I_pop_2015"}, inplace = True)

df_I_projet = pd.concat([df_I_GDP, df_I_internet, df_I_pop], axis=1, join='inner')

print(df_I_projet.head(5))
```

```python
# Création du nouvel indicateur par application de la formule

df_I_projet['I_projet_2015'] = (df_I_projet['I_GDP_2015'] * df_I_projet['I_internet_2015'] * df_I_projet['I_pop_2015'])**(1/3)
print(df_I_projet.head(5))
```

```python
# Affichage de l'indicateur projet

df_I_projet.drop(['I_GDP_2015', 'I_internet_2015', 'I_pop_2015'], axis=1, inplace=True)
```

```python
# Carte chloroplèthe de l'indicateur projet en 2015

# Fusion des deux datasets en utilisant code/iso_a3 comme identifiants uniques
for_plotting_2 = world.merge(
    df_I_projet, left_on='iso_a3', right_on='Country Code')

# Package pour aligner verticalement les axes du graphe avec celui de la légende
from mpl_toolkits.axes_grid1 import make_axes_locatable

# Création de la figure et des axes avec Matplotlib
fig, ax = plt.subplots(1, 1, figsize=(16, 16))
divider = make_axes_locatable(ax)

# Création de la colorbar
cax = divider.append_axes("right", size="5%", pad=0.1)

# Création de la carte
for_plotting_2.dropna().plot(column='I_projet_2015', cmap='YlGnBu',
                           figsize=(15, 9), ax=ax, legend=True, cax=cax)

# Titre de la carte
ax.set_title('Pays à fort potentiel en 2015', fontdict={'fontsize': 25})

# Retrait des axes
ax.set_axis_off()
```
